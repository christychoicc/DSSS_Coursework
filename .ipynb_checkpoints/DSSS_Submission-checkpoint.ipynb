Pedestrians are known as the most vulnerable road users, where they are more prone to higher injury severity levels in comparison to other road users (Pour-Rouholamin et.al., 2016). Reported by World Health Organisation (2023), more than half of global road traffic fatalities occur among vulnerable road users, including pedestrians. Pedestrian injury remains a key challenge for urban transport safety because pedestrians are especially vulnerable to severe outcomes in road collisions. In dense cities such as London, pedestrian collision severity may be shaped by a combination of road characteristics, traffic conditions, pedestrian characteristics, and wider environmental context. While many road safety studies focus on traffic engineering and demographic factors, less attention has been given to whether neighbourhood environmental features such as greenspace accessibility are associated with the severity of pedestrian injuries.

**References**

Pour-Rouholamin, M. and Zhou, H. (2016). Investigating the risk factors associated with pedestrian injury severity in Illinois. Journal of Safety Research, [online] 57, pp.9–17. doi:https://doi.org/10.1016/j.jsr.2016.03.004.

## 2. Research Question

[[ go back to the top ]](#Table-of-contents)

Pedestrian safety is a major urban policy and public health concern because severe pedestrian injuries reflect not only traffic conditions, but also the interaction between road design, pedestrian behaviour, environmental context, and neighbourhood characteristics. This study focuses on pedestrian-involved collisions in London and examines whether urban greenspace accessibility contributes meaningful explanatory value beyond established collision and casualty factors. Therefore, this analysis aims to answer the following research questions: 

**RQ1** - What factors are associated with pedestrians being killed or seriously injured rather than slightly injured in London road collisions in 2024? 

**RQ2** - Does proximity to greenspace access points contribute meaningful explanatory value to pedestrian injury severity considering collision, road, and casualty factors?

### 3.4. Data Pre-processing 

#### 3.4.1. Data Cleaning

```python
# Filter to London collisions only
collision_london = collision[
    collision['local_authority_ons_district'].astype(str).str.startswith('E09')
].copy()

print("London collisions shape:", collision_london.shape)
print(collision_london['local_authority_ons_district'].nunique())
```

```python
# Filter Greenspace data to London only
london_boundary = london_lsoa.dissolve()
# Ensure consistent CRS
greenspace = greenspace.to_crs(london_boundary.crs)
# Clip greenspace to the exact London boundary
greenspace_london = gpd.clip(greenspace, london_boundary)


# Transform geometry column's point(x y z) into normal point(x y) only
# Code derived from LLM (Claude AI)
greenspace_london['geometry'] = greenspace_london['geometry'].apply(
    lambda g: Point(g.x, g.y)  # drops Z, keeps X and Y only
)

# collision points -> nearest greenspace distance
collision_points = gpd.GeoDataFrame(
    collision_london[['collision_index', 'longitude', 'latitude']].dropna().drop_duplicates('collision_index'),
    geometry=gpd.points_from_xy(
        collision_london[['collision_index', 'longitude', 'latitude']].dropna().drop_duplicates('collision_index')['longitude'],
        collision_london[['collision_index', 'longitude', 'latitude']].dropna().drop_duplicates('collision_index')['latitude']
    ),
    crs='EPSG:4326'
).to_crs('EPSG:27700')

collision_green_nearest = gpd.sjoin_nearest(
    collision_points,
    greenspace_london[['geometry']],
    how='left',
    distance_col='dist_to_greenspace_m'
)[['collision_index', 'dist_to_greenspace_m']].drop_duplicates('collision_index')

collision_london = collision_london.merge(
    collision_green_nearest,
    on='collision_index',
    how='left'
)
```

```python
# Filter to Pedestrian casualties only
# 3 = Pedestrians
casualty_ped = casualty[casualty['casualty_class'] == 3].copy()

print("Pedestrian casualties shape:", casualty_ped.shape)
print(casualty_ped['casualty_class'].value_counts(dropna=False))
```

```python
# Filter for London Pedestrian casualties only
casualty_ped_london = casualty_ped[
    casualty_ped['collision_index'].isin(collision_london['collision_index'])
].copy()

print("London pedestrian casualties shape:", casualty_ped_london.shape)
```

```python
# Variable-specific recoding rules
recode_nan = {
    # collision
    'speed_limit': [-1, 99],
    'junction_detail': [-1, 99],
    'junction_control': [-1, 9],
    'pedestrian_crossing': [-1, 99],
    'light_conditions': [-1],
    'weather_conditions': [-1, 9],   # keep 8 = Other
    'road_surface_conditions': [-1, 9],
    'urban_or_rural_area': [-1, 3],  # 3 = Unallocated

    # casualty
    'sex_of_casualty': [-1, 9],
    'age_of_casualty': [-1],
    'age_band_of_casualty': [-1],
    'pedestrian_location': [-1],     # keep 10 for now
    'pedestrian_movement': [-1],     # keep 9 for now
    'pedestrian_road_maintenance_worker': [-1, 2],  # 2 = Not Known
    'casualty_imd_decile': [-1],

    # vehicle
    'vehicle_type': [-1],            # keep 90 = Other vehicle
    'vehicle_manoeuvre': [-1, 99],
    'sex_of_driver': [-1, 3],        # 3 = Not known
    'age_of_driver': [-1],
    'age_band_of_driver': [-1],
    'propulsion_code': [-1]
}
```

```python
# Set recode function
def apply_recode(df, recode_dict):
    df = df.copy()
    for col, bad_codes in recode_dict.items():
        if col in df.columns:
            df[col] = df[col].replace(bad_codes, np.nan)
    return df
```

```python
# Apply recode functions to the merged dataset
merged = apply_recode(merge2, recode_nan)
print("Merged and recoded clean study dataset shape:", merged.shape)
```

#### 3.4.5. Data Wrangling

The severity levels variable is defined as:

- **1 = fatal or serious pedestrian outcome**
- **0 = slight pedestrian outcome**

The analysis is conducted at the pedestrian casualty level, meaning that each row in the analytical dataset represents one pedestrian casualty involved in a road collision in London in 2024. A pedestrian casualty is encoded as 1 if their recorded injury severity is fatal or serious, and as 0 if it is slight. As the analysis is conducted at the pedestrian casualty level, some collisions may contribute more than one pedestrian collision event where multiple pedestrians injuries were involved. This is acknowledged as a limitation, as observations from the same collision event may not be entirely independent.

```python
# Create a binary for our target (severity levels)
df_clean = df_clean[df_clean['casualty_severity'].isin([1, 2, 3])].copy()
df_clean['severity_levels'] = np.where(df_clean['casualty_severity'].isin([1, 2]), 1, 0)

print(df_clean['severity_levels'].value_counts())
```

```python
# Convert time to datetime and extract hour
df_clean['time'] = pd.to_datetime(df_clean['time'], format='%H:%M', errors='coerce')
df_clean['hour'] = df_clean['time'].dt.hour
```

```python
# Drop propulsion_code (1906 missing rows)
df_clean = df_clean.drop(columns=['propulsion_code'], errors='ignore')
```

#### 3.4.6. Impute missing values before machine learning

```python
# Impute as Tree based methods cannot deal with NaNs
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Impute NaN with median
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Impute NaN with 'Missing'
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

transformed = ColumnTransformer(
    transformers=[
        # numeric
        ('num', numeric_pipeline, numeric_features),
        # categorical
        ('cat', categorical_pipeline, categorical_features)
    ]
)
```

### 4.2. Decision Tree

```python
X = df_clean[numeric_features + categorical_features].copy()
y = df_clean['severity_levels'].copy()

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval,
    test_size=0.25,   # 25% of 80% = 20% validation
    random_state=42,
    stratify=y_trainval
)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score, classification_report
)

# Decision Tree pipeline
dt_pipeline = Pipeline(steps=[
    ('transformed', transformed),
    ('model', DecisionTreeClassifier(random_state=42))
])

# Grid
param_grid_dt = {
    'model__max_depth': [3, 5, 10, 15, 20, None],
    'model__min_samples_split': [5, 10, 20],
    'model__min_samples_leaf': [1, 2, 4, 8],
    'model__criterion': ['gini', 'entropy'],
    'model__class_weight': ['balanced', {0: 1, 1: 2}, {0: 1, 1: 3}]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_dt = GridSearchCV(
    estimator=dt_pipeline,
    param_grid=param_grid_dt,
    scoring='f1',
    cv=cv,
    n_jobs=-1,
    verbose=1,
    return_train_score=True
)

grid_dt.fit(X_train, y_train)

best_dt = grid_dt.best_estimator_

print("\nBest Decision Tree hyperparameters:")
print(grid_dt.best_params_)
print(f"\nBest cross-validated F1 on training set: {grid_dt.best_score_:.4f}")

# Threshold tuning
val_probs_dt = best_dt.predict_proba(X_val)[:, 1]

thresholds_dt = np.arange(0.10, 0.91, 0.05)
threshold_rows_dt = []

for th in thresholds_dt:
    val_pred_dt = (val_probs_dt >= th).astype(int)
    threshold_rows_dt.append({
        'threshold': th,
        'precision': precision_score(y_val, val_pred_dt, zero_division=0),
        'recall': recall_score(y_val, val_pred_dt, zero_division=0),
        'f1': f1_score(y_val, val_pred_dt, zero_division=0)
    })

threshold_df_dt = pd.DataFrame(threshold_rows_dt).sort_values('f1', ascending=False)
best_threshold_dt = threshold_df_dt.iloc[0]['threshold']

print(f"\nSelected threshold based on validation F1: {best_threshold_dt:.2f}")
print(threshold_df_dt.head())

# Refit on train validation
best_dt.fit(X_trainval, y_trainval)



# Test evaluation
test_probs_dt = best_dt.predict_proba(X_test)[:, 1]
test_pred_dt = (test_probs_dt >= best_threshold_dt).astype(int)

test_accuracy_dt = accuracy_score(y_test, test_pred_dt)
test_precision_dt = precision_score(y_test, test_pred_dt, zero_division=0)
test_recall_dt = recall_score(y_test, test_pred_dt, zero_division=0)
test_f1_dt = f1_score(y_test, test_pred_dt, zero_division=0)
test_roc_auc_dt = roc_auc_score(y_test, test_probs_dt)
test_pr_auc_dt = average_precision_score(y_test, test_probs_dt)

print("\nDecision Tree Test Results")
print(f"Accuracy : {test_accuracy_dt:.4f}")
print(f"Precision: {test_precision_dt:.4f}")
print(f"Recall   : {test_recall_dt:.4f}")
print(f"F1-score : {test_f1_dt:.4f}")
print(f"ROC-AUC  : {test_roc_auc_dt:.4f}")
print(f"PR-AUC   : {test_pr_auc_dt:.4f}\n")

print("Decision Tree Classification report:")
print(classification_report(
    y_test,
    test_pred_dt,
    target_names=['Slight', 'Fatal/ Serious']
))

dt_results = {
    # 'Train_Share': X_train.shape[0] / len(X),
    # 'Validation_Share': X_val.shape[0] / len(X),
    # 'Test_Share': X_test.shape[0] / len(X),
    # 'Best_CV_F1': grid_dt.best_score_,
    # 'Best_Threshold': best_threshold_dt,
    'Test_Accuracy': test_accuracy_dt,
    'Test_Precision': test_precision_dt,
    'Test_Recall': test_recall_dt,
    'Test_F1': test_f1_dt,
    'Test_ROC_AUC': test_roc_auc_dt,
    'Test_PR_AUC': test_pr_auc_dt
}

print(dt_results)
```

## 4.4. SHAP Model Interpretation

```python
import shap
import numpy as np

# for faster plotting, sample test data
X_test_sample = X_test.sample(n=min(500, len(X_test)), random_state=42)

# use the fitted preprocessing step and fitted RF from your best pipeline
preprocessor = best_rf.named_steps['transformed']
rf_model = best_rf.named_steps['model']

# transform test data into the same format seen by the RF
X_test_sample_transformed = preprocessor.transform(X_test_sample)

# convert sparse matrix to dense if needed
if hasattr(X_test_sample_transformed, "toarray"):
    X_test_sample_transformed = X_test_sample_transformed.toarray()

# force numeric dtype
X_test_sample_transformed = np.asarray(X_test_sample_transformed, dtype=np.float64)

# get transformed feature names
feature_names = preprocessor.get_feature_names_out()
feature_names = [f.replace("num__", "").replace("cat__", "") for f in feature_names]

# SHAP explainer
explainer = shap.TreeExplainer(rf_model)
shap_values = explainer(X_test_sample_transformed)

print("SHAP values shape:", shap_values.values.shape)

# keep class 1 = fatal/serious
shap_values_class1 = shap.Explanation(
    values=shap_values.values[:, :, 1],
    base_values=shap_values.base_values[:, 1],
    data=shap_values.data,
    feature_names=feature_names
)

print("Class 1 SHAP shape:", shap_values_class1.values.shape)
```

```python
# beeswarm plot
shap.plots.beeswarm(shap_values_class1, max_display=15)
```

Greenspace accessibility contributed lightly on explanatory value for severe pedestrian injury when pedestrian, road, collision, vehicle, and driver variables were accounted for. SHAP analysis identified casualty age, junction-related variables, vehicle manoeuvre, driver characteristics, and road context as the primary predictors. Distance to the nearest greenspace access point was retained in the final random forest model but carried a low mean absolute SHAP value of 0.0031, indicating that it contributes relatively less in comparison to the characteristics of the collision environment and individuals involved.

## 5. Results and discussion

[[ go back to the top ]](#Table-of-contents)

The Random Forest model outperformed the Decision Tree model, achieving an accuracy of 0.55 and a recall of 0.78 for fatal/ serious severity levels. Under a road saftey context, a higher recall score is recognised to be more important metric, where this model is able to identify 78% of actual fatal/ serious injuries. A higher recall score also comes with a trade off of a lower precision score. The model achieved a 0.35 precision score, suggesting that the model tends to over predict severity levels.  

**RQ1:** SHAP analysis of the weighted Random Forest identified the most featured importance predictors. Casualty age is found as the most influential variable, consistent with greater physiological vulnerability among older pedestrians, who are substantially more likely to acquire fatal or serious outcomes. Junction-related variables ranked highly with junction_detail and junction_location, reflecting that the presence of concentration of severe collisions at complex navigation points where vehicle movements may collide with pedestrian usages. Vehicle manoeuvre and vehicle type also carried meaningful explanatory weight, positioning the importance of collision physics in determining severity outcome. Driver-level demographic variables, including age_of_driver and sex_of_driver contributed additional explanatory value, suggesting that driver behaviour and reaction capacity are also relevant factors. Road environment variables, such as speed limit and pedestrian crossing type are featured in the at lower parts of the SHAP rankings, indicating a secondary but a substaintial structural influence on severity.

**RQ2:** Distance to the nearest greenspace access point was preserved in the main Random Forest model as a spatial contextual variable. Its mean absolute SHAP value was found as 0.0031, placing it at a lower position as a predictor    in the model. Once pedestrian, junction, vehicle, and driver characteristics were accounted for, greenspace proximity contributed negligible additional explanatory value. This suggests that, within the analytical framework applied here, the immediate physical and human characteristics of the collision environment are far stronger determinants of injury severity than the surrounding spatial context.

## 6. Conclusion

[[ go back to the top ]](#Table-of-contents)

Pedestrian age and junction complexity were the most important predictors of severity levels, reflecting both biological vulnerability and structural risk of road users. The model achieved a recall of 78%, making it a viable screening tool for transport planners seeking to identify potentially lethal road configurations, accepting lower precision in exchange for greater sensitivity. Despite greenspace accessibility to be relevant to wider urban health outcomes, yet it did not emerge as a primary driver of injury severity when modelled against conventional traffic engineering factors.

These findings also has direct policy relevance, supporting the "Vision Zero" strategy in London's transport system, particularly speed reduction and improved junction visibility, especially in areas with higher concentrations of higher aged pedestrian road users, where such measures carry greater weight than wider environmental features in determining collision outcomes.
