What does this study aim to do --> bridge the gap. 

Pedestrians are known as the most vulnerable road users, where they are more prone to higher injury severity levels in comparison to other road users (Pour-Rouholamin et.al., 2016). Reported by World Health Organisation (2023), more than half of global road traffic fatalities occur among vulnerable road users, including pedestrians. Pedestrian injury remains a key challenge for urban transport safety because pedestrians are especially vulnerable to severe outcomes in road collisions. In dense cities such as London, pedestrian collision severity may be shaped by a combination of road characteristics, traffic conditions, pedestrian characteristics, and wider environmental context. While many road safety studies focus on traffic engineering and demographic factors, less attention has been given to whether neighbourhood environmental features such as greenspace accessibility are associated with the severity of pedestrian injuries.

**References**

Pour-Rouholamin, M. and Zhou, H. (2016). Investigating the risk factors associated with pedestrian injury severity in Illinois. Journal of Safety Research, [online] 57, pp.9–17. doi:https://doi.org/10.1016/j.jsr.2016.03.004.

World Health Organization (2023). Road traffic injuries. [online] World Health Organization. Available at: https://www.who.int/news-room/fact-sheets/detail/road-traffic-injuries.

## 2. Research Question

[[ go back to the top ]](#Table-of-contents)

Pedestrian safety is a major urban policy and public health concern because severe pedestrian injuries reflect not only traffic conditions, but also the interaction between road design, pedestrian behaviour, environmental context, and neighbourhood characteristics. This study focuses on pedestrian-involved collisions in London and examines whether urban greenspace accessibility contributes meaningful explanatory value beyond established collision and casualty factors. Therefore, this analysis aims to answer the following research questions: 

**RQ1** - What factors are associated with pedestrians being killed or seriously injured rather than slightly injured in London road collisions in 2024? 

**RQ2** - Which factors are most important in explaining why some pedestrian collisions lead to more severe injuries?

### 3.4. Data Pre-processing 

#### 3.4.1. Data Cleaning

```python
# Filter to London collisions only
collision_london = collision[
    collision['local_authority_ons_district'].astype(str).str.startswith('E09')
].copy()

print("London collisions shape:", collision_london.shape)
print(collision_london['local_authority_ons_district'].nunique())
```

```python
# Filter to Pedestrian casualties only
# 3 = Pedestrians
casualty_ped = casualty[casualty['casualty_class'] == 3].copy()

print("Pedestrian casualties shape:", casualty_ped.shape)
print(casualty_ped['casualty_class'].value_counts(dropna=False))
```

```python
# Filter for London Pedestrian casualties only
casualty_ped_london = casualty_ped[
    casualty_ped['collision_index'].isin(collision_london['collision_index'])
].copy()

print("London pedestrian casualties shape:", casualty_ped_london.shape)
```

#### 3.4.2. Data Joins/ Merge

```python
# Merge collision and casualty data on collision_index
merge = casualty_ped_london.merge(
    collision_london,
    on='collision_index',
    how='left',
    suffixes=('_cas', '_col')
)

print("After collision casualty merge:", merge.shape)
```

```python
# Merge collision and casualty data with vehicle data
# Drop columns unneeded columns that may clash
vehicle_merge = vehicle.drop(
    columns=['collision_year', 'collision_ref_no'],
    errors='ignore'
)

merge2 = merge.merge(
    vehicle_merge,
    on=['collision_index', 'vehicle_reference'],
    how='left'
)

print("After vehicle merge:", merge2.shape)
```

#### 3.4.3. Data Ambiguity (Missing, Unknown, Undefined, Others)

```python
# Variable-specific recoding rules
recode_nan = {
    # collision
    'speed_limit': [-1, 99],
    'junction_detail': [-1, 99],
    'junction_control': [-1, 9],
    'pedestrian_crossing': [-1, 99],
    'light_conditions': [-1],
    'weather_conditions': [-1, 9],   # keep 8 = Other
    'road_surface_conditions': [-1, 9],
    'urban_or_rural_area': [-1, 3],  # 3 = Unallocated

    # casualty
    'sex_of_casualty': [-1, 9],
    'age_of_casualty': [-1],
    'age_band_of_casualty': [-1],
    'pedestrian_location': [-1],     # keep 10 for now
    'pedestrian_movement': [-1],     # keep 9 for now
    'pedestrian_road_maintenance_worker': [-1, 2],  # 2 = Not Known
    'casualty_imd_decile': [-1],

    # vehicle
    'vehicle_type': [-1],            # keep 90 = Other vehicle
    'vehicle_manoeuvre': [-1, 99],
    'sex_of_driver': [-1, 3],        # 3 = Not known
    'age_of_driver': [-1],
    'age_band_of_driver': [-1],
    'propulsion_code': [-1]
}
```

```python
# Set recode function
def apply_recode(df, recode_dict):
    df = df.copy()
    for col, bad_codes in recode_dict.items():
        if col in df.columns:
            df[col] = df[col].replace(bad_codes, np.nan)
    return df
```

```python
# Apply recode functions to the merged dataset
merged = apply_recode(merge2, recode_nan)
print("Merged and recoded clean study dataset shape:", merged.shape)
```

#### 3.4.4. Data Filtering

#### 3.4.5. Data Wrangling

```python
# Create a binary for our target (severity levels)
df_clean = df_clean[df_clean['casualty_severity'].isin([1, 2, 3])].copy()
df_clean['severity_levels'] = np.where(df_clean['casualty_severity'].isin([1, 2]), 1, 0)

print(df_clean['severity_levels'].value_counts())
```

```python
# Convert time to datetime and extract hour
df_clean['time'] = pd.to_datetime(df_clean['time'], format='%H:%M', errors='coerce')
df_clean['hour'] = df_clean['time'].dt.hour
```

```python
# Drop propulsion_code (1906 missing rows)
df_clean = df_clean.drop(columns=['propulsion_code'], errors='ignore')
```

```python
# Define features
numeric_features = [
    'age_of_casualty',
    'longitude',
    'latitude',
    'number_of_vehicles',
    'age_of_driver',
    'age_of_vehicle',
    'engine_capacity_cc',
    'hour'
]
categorical_features = [
    'sex_of_casualty',
    'pedestrian_location',
    'pedestrian_movement',
    'pedestrian_road_maintenance_worker',
    'casualty_type',
    'casualty_imd_decile',
    'day_of_week',
    'speed_limit',
    'first_road_class',
    'road_type',
    'junction_detail',
    'junction_control',
    'pedestrian_crossing',
    'light_conditions',
    'weather_conditions',
    'road_surface_conditions',
    'urban_or_rural_area',
    'trunk_road_flag',
    'vehicle_type',
    'vehicle_manoeuvre',
    'junction_location',
    'sex_of_driver',
    'driver_imd_decile',
    'escooter_flag'
]
```

```python
# Assigne categroical features as objects (aka categories make them not comparable numbers)
import numpy as np
import numpy as np
import pandas as pd

# Numeric columns: keep numeric
for col in numeric_features:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# Categorical columns: convert all non-missing values to strings, keep missing as np.nan
for col in categorical_features:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].apply(lambda x: np.nan if pd.isna(x) else str(x))
```

```python
print("Wrangled dataset shape:", df_clean.shape)
print(df_clean.columns.tolist())
```

```python
missing_unknown_variables = {
    # collision
    'speed_limit': [-1, 99],
    'junction_detail': [-1, 99],
    'junction_control': [-1, 9],
    'pedestrian_crossing': [-1, 99],
    'light_conditions': [-1],
    'weather_conditions': [-1, 9],   # 8 kept as Other
    'road_surface_conditions': [-1, 9],
    'urban_or_rural_area': [-1, 3],

    # casualty
    'sex_of_casualty': [-1, 9],
    'age_of_casualty': [-1],
    'age_band_of_casualty': [-1],
    'pedestrian_location': [-1],
    'pedestrian_movement': [-1],
    'pedestrian_road_maintenance_worker': [-1, 2],
    'casualty_imd_decile': [-1],

    # vehicle
    'vehicle_type': [-1],
    'vehicle_manoeuvre': [-1, 99],
    'sex_of_driver': [-1, 3],
    'age_of_driver': [-1],
    'age_band_of_driver': [-1],
    'propulsion_code': [-1]
}

# Check NaN counts in the variables
for col in missing_unknown_variables.keys():
    if col in df_clean.columns:
        missing_count = df_clean[col].isna().sum()
        missing_pct = round(df_clean[col].isna().mean() * 100, 2)
        print(f"{col}: NaN count = {missing_count}, NaN % = {missing_pct}")
```

#### 3.4.6. Impute missing values before machine learning

```python
# Impute as Tree based methods cannot deal with NaNs
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Impute NaN with median
numeric_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Impute NaN with 'Missing'
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

transformed = ColumnTransformer(
    transformers=[
        # numeric
        ('num', numeric_pipeline, numeric_features),
        # categorical
        ('cat', categorical_pipeline, categorical_features)
    ]
)
```

### 4.2. Decision Tree

```python
X = df_clean[numeric_features + categorical_features].copy()
y = df_clean['severity_levels'].copy()

X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval,
    test_size=0.25,   # 25% of 80% = 20% validation
    random_state=42,
    stratify=y_trainval
)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score, classification_report
)

# Decision Tree pipeline
dt_pipeline = Pipeline(steps=[
    ('transformed', transformed),
    ('model', DecisionTreeClassifier(random_state=42))
])

# Grid
param_grid_dt = {
    'model__max_depth': [3, 5, 10, 15, 20, None],
    'model__min_samples_split': [5, 10, 20],
    'model__min_samples_leaf': [1, 2, 4, 8],
    'model__criterion': ['gini', 'entropy'],
    'model__class_weight': ['balanced', {0: 1, 1: 2}, {0: 1, 1: 3}]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_dt = GridSearchCV(
    estimator=dt_pipeline,
    param_grid=param_grid_dt,
    scoring='f1',
    cv=cv,
    n_jobs=-1,
    verbose=1,
    return_train_score=True
)

grid_dt.fit(X_train, y_train)

best_dt = grid_dt.best_estimator_

print("\nBest Decision Tree hyperparameters:")
print(grid_dt.best_params_)
print(f"\nBest cross-validated F1 on training set: {grid_dt.best_score_:.4f}")

# Threshold tuning
val_probs_dt = best_dt.predict_proba(X_val)[:, 1]

thresholds_dt = np.arange(0.10, 0.91, 0.05)
threshold_rows_dt = []

for th in thresholds_dt:
    val_pred_dt = (val_probs_dt >= th).astype(int)
    threshold_rows_dt.append({
        'threshold': th,
        'precision': precision_score(y_val, val_pred_dt, zero_division=0),
        'recall': recall_score(y_val, val_pred_dt, zero_division=0),
        'f1': f1_score(y_val, val_pred_dt, zero_division=0)
    })

threshold_df_dt = pd.DataFrame(threshold_rows_dt).sort_values('f1', ascending=False)
best_threshold_dt = threshold_df_dt.iloc[0]['threshold']

print(f"\nSelected threshold based on validation F1: {best_threshold_dt:.2f}")
print(threshold_df_dt.head())

# Refit on train validation
best_dt.fit(X_trainval, y_trainval)



# Test evaluation
test_probs_dt = best_dt.predict_proba(X_test)[:, 1]
test_pred_dt = (test_probs_dt >= best_threshold_dt).astype(int)

test_accuracy_dt = accuracy_score(y_test, test_pred_dt)
test_precision_dt = precision_score(y_test, test_pred_dt, zero_division=0)
test_recall_dt = recall_score(y_test, test_pred_dt, zero_division=0)
test_f1_dt = f1_score(y_test, test_pred_dt, zero_division=0)
test_roc_auc_dt = roc_auc_score(y_test, test_probs_dt)
test_pr_auc_dt = average_precision_score(y_test, test_probs_dt)

print("\nDecision Tree Test Results")
print(f"Accuracy : {test_accuracy_dt:.4f}")
print(f"Precision: {test_precision_dt:.4f}")
print(f"Recall   : {test_recall_dt:.4f}")
print(f"F1-score : {test_f1_dt:.4f}")
print(f"ROC-AUC  : {test_roc_auc_dt:.4f}")
print(f"PR-AUC   : {test_pr_auc_dt:.4f}\n")

print("Decision Tree Classification report:")
print(classification_report(
    y_test,
    test_pred_dt,
    target_names=['Slight', 'Fatal/ Serious']
))

dt_results = {
    'Train_Share': X_train.shape[0] / len(X),
    'Validation_Share': X_val.shape[0] / len(X),
    'Test_Share': X_test.shape[0] / len(X),
    'Best_CV_F1': grid_dt.best_score_,
    'Best_Threshold': best_threshold_dt,
    'Test_Accuracy': test_accuracy_dt,
    'Test_Precision': test_precision_dt,
    'Test_Recall': test_recall_dt,
    'Test_F1': test_f1_dt,
    'Test_ROC_AUC': test_roc_auc_dt,
    'Test_PR_AUC': test_pr_auc_dt
}

print(dt_results)
```

### 4.3. Random Forest

```python
# import numpy as np
# import pandas as pd
# import matplotlib.pyplot as plt

# from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
# from sklearn.pipeline import Pipeline
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import (
#     classification_report,
#     accuracy_score, precision_score, recall_score, f1_score,
#     roc_auc_score, average_precision_score,
#     ConfusionMatrixDisplay
# )

X = df_clean[numeric_features + categorical_features].copy()
y = df_clean['severity_levels'].copy()

print("X shape:", X.shape)
print("y distribution:")
print(y.value_counts())
print(y.value_counts(normalize=True).round(3))

# Train Test Split
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval,
    test_size=0.25,   # 25% of 80% = 20% of full data
    random_state=42,
    stratify=y_trainval
)

print("Train:", X_train.shape, y_train.shape)
print("Validation:", X_val.shape, y_val.shape)
print("Test:", X_test.shape, y_test.shape)


rf_pipeline = Pipeline(steps=[
    ('transformed', transformed),
    ('model', RandomForestClassifier(
        random_state=42,
        n_jobs=-1
    ))
])


# Set prarms
param_grid_rf = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [5, 10, 15, None],
    'model__min_samples_leaf': [1, 2, 4],
    'model__max_features': ['sqrt', 'log2'],
    'model__class_weight': ['balanced', {0: 1, 1: 2}, {0: 1, 1: 3}]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_rf = GridSearchCV(
    estimator=rf_pipeline,
    param_grid=param_grid_rf,
    scoring='f1', # choose f1 cus data imbalance
    cv=cv,
    n_jobs=-1,
    verbose=1,
    return_train_score=True
)


grid_rf.fit(X_train, y_train)

best_rf = grid_rf.best_estimator_

print("\nBest hyperparameters:")
print(grid_rf.best_params_)

print(f"\nBest cross-validated F1 on training set: {grid_rf.best_score_:.4f}")


# Threshold tuning
val_probs_rf = best_rf.predict_proba(X_val)[:, 1]

thresholds_rf = np.arange(0.10, 0.91, 0.05)
threshold_rows_rf = []

for th in thresholds_rf:
    val_pred_rf = (val_probs_rf >= th).astype(int)
    threshold_rows_rf.append({
        'threshold': th,
        'precision': precision_score(y_val, val_pred_rf, zero_division=0),
        'recall': recall_score(y_val, val_pred_rf, zero_division=0),
        'f1': f1_score(y_val, val_pred_rf, zero_division=0)
    })

threshold_df_rf = pd.DataFrame(threshold_rows_rf).sort_values('f1', ascending=False)

best_threshold_rf = threshold_df_rf.iloc[0]['threshold']

print(f"\nSelected threshold based on validation F1: {best_threshold_rf:.2f}")

# Best 
best_rf.fit(X_trainval, y_trainval)

test_probs_rf = best_rf.predict_proba(X_test)[:, 1]
test_pred_rf = (test_probs_rf >= best_threshold_rf).astype(int)

test_accuracy_rf = accuracy_score(y_test, test_pred_rf)
test_precision_rf = precision_score(y_test, test_pred_rf, zero_division=0)
test_recall_rf = recall_score(y_test, test_pred_rf, zero_division=0)
test_f1_rf = f1_score(y_test, test_pred_rf, zero_division=0)
test_roc_auc_rf = roc_auc_score(y_test, test_probs_rf)
test_pr_auc_rf = average_precision_score(y_test, test_probs_rf)

print("\nRandom Forest Test Results")
print(f"Accuracy : {test_accuracy_rf:.4f}")
print(f"Precision: {test_precision_rf:.4f}")
print(f"Recall   : {test_recall_rf:.4f}")
print(f"F1-score : {test_f1_rf:.4f}")
print(f"ROC-AUC  : {test_roc_auc_rf:.4f}")
print(f"PR-AUC   : {test_pr_auc_rf:.4f}\n")

print("Reandom Forest Classification report:")
print(classification_report(
    y_test,
    test_pred_rf,
    target_names=['Slight', 'Fatal/ Serious']
))


# Save as a dictionary
rf_results = {
    'Train_Share': X_train.shape[0] / len(X),
    'Validation_Share': X_val.shape[0] / len(X),
    'Test_Share': X_test.shape[0] / len(X),
    'Best_CV_F1': grid_rf.best_score_,
    'Best_Threshold': best_threshold_rf,
    'Test_Accuracy': test_accuracy_rf,
    'Test_Precision': test_precision_rf,
    'Test_Recall': test_recall_rf,
    'Test_F1': test_f1_rf,
    'Test_ROC_AUC': test_roc_auc_rf,
    'Test_PR_AUC': test_pr_auc_rf
}
```

## 4. Methodology

[[ go back to the top ]](#Table-of-contents)

**Methodology Justification**

This study used a 4 methodological approach. Exploratory analysis was first used to describe the distribution of pedestrian collision severity and greenspace accessibility. Moran’s I was applied as an exploratory spatial diagnostic to assess whether LSOA level severity rates exhibits spatial clustering across London. For modelling, a Decision Tree classifier was used as a baseline model because due to its transparency and interpretability. A Random Forest classifier was then selected as the main model as it can capture non-linear relationships and interactions between pedestrian, road, environmental, and spatial variables more effectively than a single decision tree. SHAP values were then used to assess the relative importance of greenspace accessibility alongside other predicting factors such as age_of_casualty, etc.






This study uses a staged analytical approach designed to examine factors associated with fatal or serious pedestrian collision outcomes in London while keeping the analysis aligned with the structure of the available data. First, exploratory data analysis is used to summarise the distribution of pedestrian collision severity, inspect the distribution of greenspace accessibility, and identify broad patterns in the data before modelling. This step is important because it provides descriptive context for the later analysis and helps assess whether simple patterns are visible prior to applying machine-learning methods.

Second, an exploratory spatial analysis is conducted using choropleth mapping and Moran’s I. These methods are used to examine whether area-level severity rates show broad spatial clustering across London LSOAs. This spatial step is included to provide geographic context for the study, rather than to establish causality or determine model choice. In other words, the spatial analysis helps describe whether severe outcomes appear to cluster across space, but it is not treated as a substitute for collision-level modelling.

Third, a Decision Tree classifier is used as a simple baseline model. The value of the Decision Tree is that it provides an interpretable benchmark and allows the classification problem to be approached using a method that is relatively transparent. However, because pedestrian collision severity may depend on non-linear relationships and interactions between pedestrian, road, and environmental characteristics, a Random Forest classifier is then used as the main model. Random Forest is appropriate here because it can capture more complex relationships than a single tree and is robust to mixed predictor types after encoding.

Finally, SHAP is used to interpret the Random Forest model. This is especially useful for the second research question, which asks how important distance to the nearest greenspace access point is relative to other factors. SHAP provides a transparent way to examine the relative importance of each variable within the fitted model and therefore helps assess whether greenspace accessibility contributes meaningfully alongside more established predictors such as pedestrian age, speed limit, and road characteristics.

### 4.2. Tree-based Methods

**Variables Selection**

The predictor set was designed to include variables that plausibly describe the pedestrian, road, environmental, and spatial context of each collision. These include pedestrian characteristics such as age, sex, movement, and location; road and traffic characteristics such as speed limit, road type, junction detail, junction control, crossing facilities, and number of vehicles involved; environmental conditions such as lighting, weather, and road surface conditions; and the spatial measure of distance to the nearest greenspace access point.

number_of_vehicles was retained because it reflects the complexity of the collision context and may be associated with greater conflict intensity or more severe impact conditions. By contrast, number_of_casualties was excluded from the final model because it is too closely related to the outcome itself and may therefore blur the distinction between explanatory variables and collision consequences. Removing this variable improves the conceptual clarity of the modelling framework, even though it may slightly reduce predictive performance.


**Imbalance Data Approach**

The outcome variable is imbalanced, with fatal or serious pedestrian collisions forming the minority class and slight collisions forming the majority class. Although this imbalance is not extreme, it is still important because standard classifiers may otherwise become biased toward the majority class and perform poorly in identifying severe cases.

To address this issue, three steps were taken. First, stratified sampling was used when splitting the data into training, validation, and test sets so that the proportion of severe and slight collisions remained consistent across all subsets. Second, class weights were incorporated into the Random Forest tuning process so that misclassification of the minority fatal/serious class received greater penalty during model fitting. Third, the classification threshold was selected using the validation set rather than assuming the default probability threshold would be optimal. The threshold was chosen based on F1-score in order to balance precision and recall for the minority severe class.

This approach was preferred to more aggressive resampling methods because it keeps the observed data structure intact while still responding to imbalance in a transparent and defensible way.
